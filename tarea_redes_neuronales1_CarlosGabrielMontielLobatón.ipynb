Problema 1: Proceso dicotómico de Markov (Proceso telegráfico aleatorio)

la función $Y(t)$ es un proceso de Markov entre {-1 y 1} que conmuta entre los valores de la tasa $γ$

a) contruir el ecuación maestra el proceso dicotomico de Marcok

b) Resolver la ecuación maestra cuya condición inicial es $Y(t_0) = y_0$ para obtener la densidad de probabilidad condicional $P(y,t ¦ y_0,t_0)$



**SOLUCIÓN**

*A)*
vea que:

$P(1,t)$ es la probabilidad de $Y(t) = -1$ en un tiempo $t$

$P(-1,t)$ es la probabilidad de $Y(t) = 1$ en un tiempo $t$


**Ecuación maestra**

$\frac{d}{dt}P(1,t) = \gamma P(-1,t) - \gamma P(1,t)$

$\frac{d}{dt}P(-1,t) = \gamma P(1,t) - \gamma P(-1,t)$

esto es que de manera matricial:

$\frac{d}{dt} \begin{pmatrix} P(1,t)\\P(-1,t) \end{pmatrix} = \begin{pmatrix} -γ, γ\\γ,-γ \end{pmatrix} \begin{pmatrix}  P(1,t)\\P(-1,t) \end{pmatrix}$

*B)* note qué:

$ν(t) = \sum_{y} yP(y,t) = 1P(1,t) + (-1)P(-1,t) = P(1,t) - P(-1,t)$

$→ \frac{d}{dt}ν(t) = \frac{d}{dt}P(1,t) - \frac{d}{dt}P(-1,t) $

$→ \frac{d}{dt}ν(t) = (\gamma P(-1,t) - \gamma P(1,t))  - (\gamma P(1,t) - \gamma P(-1,t))$
$→ \frac{d}{dt}ν(t) = -2γ(P(1,t) - P(-1,t)) = -2γν $

así obtenemos la ecuación diferencial:

$ν' + 2νγ = 0,$

de la cual la solución es una exponencial fácil de encontrar, y al no partír del tiempo cero, sino de un tiempo $t_0$ arbitrario es dadas las condiciones iniciales:

$ ν(t) = y_0 e^{-2γ(t-t_0)}$

ahora vea que

$ν(t) = P(1,t) - P(-1,t)$ y  $P(1,t) + P(1,t) = 1$

$→ P(1,t) = \frac{1}{2}(1+ν(t))$ y $P(-1,t) = \frac{1}{2}(1-ν(t))$

del cual ya es inmediato notar que:

$P(y,t¦y_0,t_0) = \frac{1}{2}(1 + yν(t)) = \frac{1}{2}(1 + yy_0 e^{-2γ(t-t_0)})$







Problema 2: Sistema multiestable (Toggle switch)

Una red de "toggle switch" consiste en dos especies que se reprimen mutuamente, $r_1$ y $r_2$. Si $r_1$ es alto, la síntesis de $r_2$ es baja y, recíprocamente, si $r_2$ es alto, $r_1$ se mantiene bajo. Una red simple que describe este sistema está dada por las siguientes cuatro reacciones:

$r_1 \xrightarrow{\nu_1} r_1 + 1, \quad \nu_1 = \alpha \, g_R(r_2/\Omega)$


$r_2 \xrightarrow{\nu_2} r_2 + 1, \quad \nu_2 = \alpha \, g_R(r_1/\Omega)$

$r_1 \xrightarrow{\nu_3} r_1 - 1, \quad \nu_3 = \beta \frac{r_1}{\Omega}$

$r_2 \xrightarrow{\nu_4} r_2 - 1, \quad \nu_4 = \beta \frac{r_2}{\Omega}$

donde la función $g_R(x)$ es alta si $x$ es bajo, y baja si $x$ es alto.

**a)** Escribe la matriz de estequiometría $\mathbf{S}$ y el vector de propensiones $\vec{\nu}$ para el toggle switch. Escribe las ecuaciones deterministas de tasas de reacción y la ecuación maestra.

**b)** Supón que $g_R(x)$ toma la forma de Hill:


$g_R(x) = \frac{1}{1 + \left( \frac{x}{K_R} \right)^n}$,

donde $K_R$ mide la fuerza del represor (cuanto menor es $K_R$, menos represor se necesita para reducir $g_R$) y $n$ es la cooperatividad que determina cuán abrupta es la transición del estado alto al bajo. Realiza simulaciones estocásticas del modelo ($\alpha = 2$, $\beta = 1$, $K_R = 1$, $n = 2$) para distintos tamaños del sistema $\Omega$. ¿Observaste conmutación espontánea en las simulaciones estocásticas?


**SOLUCIÓN**

*A)*
si $R = \begin{pmatrix} r_1\\r_2 \end{pmatrix}$

entonces: $S = \begin{pmatrix} 1,0,-1,0\\0,1,0,-1 \end{pmatrix} $

el vector de propensiones
${\nu(t)} = \begin{pmatrix} αg_{R}(\frac{r2}{Ω})\\αg_{R}(\frac{r1}{Ω})\\β\frac{r1}{Ω}\\β\frac{r2}{Ω} \end{pmatrix} $

usamos que $\frac{d}{dt}R = S\cdotν(t)$, de ello se obtiene el que:

$\frac{dr_1}{dt} = αg_{R}(\frac{r2}{Ω}) - β\frac{r1}{Ω}$ y $\frac{dr_2}{dt} = αg_{R}(\frac{r1}{Ω}) - β\frac{r2}{Ω} $

así dado que $x_i = \frac{ri}{Ω}$ recordando que $Ω$ es el tamño del sistema,

de este modo:

$\frac{dx_1}{dt} = αg_{R}(x_2) - βx_1$ y $\frac{dx_2}{dt} = αg_{R}(x_1) - βx_2 $


finalmente escribimos la ecuación maestra:

$\frac{dP(r_1,r_2,t)}{dt}=\alpha \, g_R(r_2/\Omega)P(r_1-1,r_2,t)+\alpha \, g_R(r_1/\Omega)P(r_1,r_2-1,t)+\beta \frac{r_1+1}{\Omega}P(r_1+1,r_2,t)+\beta \frac{r_2+1}{\Omega}P(r_1,r_2+1,t)-[\alpha \, g_R(r_2/\Omega)+\alpha g_R(r_1/\Omega)+\beta \frac{r_1}{\Omega}+\beta \frac{r_2}{\Omega}]P(r_1,r_2,t)$

```python
#Librerias que vamos a usar
import numpy as np
import matplotlib.pyplot as plt
import random as rnd
```

```python
#condiciones que da el problema como datos a usar
alpha = 2
betha = 1
KR = 1
n = 2
Tmax = 500
```

```python
# Función de Hill
def gR(x):
    return 1.0 / (1.0 + (x / KR)**n)

# Inicializamos
t = 0.0
r1 = 10
r2 = 50
Omega = 100

#Parametros
times = [t]
r1_traj = [r1]
r2_traj = [r2]


# Algoritmo de Gillespie con Omegas aleatorios
from tqdm import tqdm
for i in tqdm(range(20), desc="Simulando"):
  a = np.random.rand()
  b = np.random.rand()
  Omega = a*Omega/b
  while t < Tmax:

      # Propensiones
      a1 = alpha*gR(r2 / Omega)        # Producción r1
      a2 = alpha*gR(r1 / Omega)        # Producción r2
      a3 = betha*(r1 / Omega)           # Degradación r1
      a4 = betha*(r2 / Omega)           # Degradación r2
      a0 = a1 + a2 + a3 + a4

      if a0 == 0:
          break

      # Tiempo hasta siguiente reacción
      tau = -np.log(np.random.rand()) / a0
      t += tau

      # Elegir reacción
      r = np.random.rand() * a0

      if r < a1:
          r1 += 1
      elif r < a1 + a2:
          r2 += 1
      elif r < a1 + a2 + a3:
          if r1 > 0:
              r1 -= 1
      else:
          if r2 > 0:
              r2 -= 1

      times.append(t)
      r1_traj.append(r1)
      r2_traj.append(r2)

# Gráficamos (dentro del mismo ciclo para gráficar de una todos los Omegas)
  plt.figure()
  plt.plot(times, r1_traj, label="r1")
  plt.plot(times, r2_traj, label="r2")
  plt.xlabel("Tiempo")
  plt.ylabel("Número de moléculas")
  plt.legend()
  plt.title(f"Toggle switch estocástico (Omega = {Omega})")
  plt.show()
```

Problema 3: Movimiento browniano en un campo magnético

Una partícula browniana de masa $m$ y carga $q$ se mueve en el espacio tridimensional bajo la influencia de un campo magnético constante y homogéneo $\vec{B}$ dirigido a lo largo del eje $z$: $\vec{B} = (0, 0, B)$. El vector velocidad $\vec{V}(t) = (V_x(t), V_y(t), V_z(t))^T$ satisface la ecuación de Langevin con tres componentes:


$m \frac{d\vec{V}}{dt} = -\beta \vec{V}(t) - q \, \vec{B} \times \vec{V}(t) + \sqrt{2\beta k_B T} \, \vec{\Gamma}(t)$,

donde $\beta$ es el coeficiente de fricción y $\vec{\Gamma}(t) = (\Gamma_x(t), \Gamma_y(t), \Gamma_z(t))^T$ es un vector de ruido blanco gaussiano con componentes independientes.

**(a)** Simula la dinámica de la velocidad.

```python
#Librerias que vamos a usar (otra vez xd)
import numpy as np
import matplotlib.pyplot as plt
import random as rnd
```

```python
# Condiciones dadas por le problema
m = 1.1
beta = 0.5
q = 1.2
B = 2.0
kBT = 10

# Parámetros de simulación
dt = 0.0001
T_total = 100.0
n_steps = int(T_total / dt)

# Vector de tiempos
t = np.linspace(0, T_total, n_steps)

# Arrays para almacenar las velocidades
vx = np.zeros(n_steps)
vy = np.zeros(n_steps)
vz = np.zeros(n_steps)

# Condiciones iniciales
vx[0] = 0.0
vy[0] = 0.0
vz[0] = 1.0  # Velocidad inicial en z para ver mejor el efecto
```

```python
# Factor para el ruido térmico (relación de fluctuación-disipación)
D = np.sqrt(2 * beta * kBT * dt / m)

# Barra de progreso simple
from tqdm import tqdm
for i in tqdm(range(n_steps - 1), desc="Simulando"):
    # Generar ruido blanco gaussiano para cada componente
    Gamma_x = np.random.normal(0, D)
    Gamma_y = np.random.normal(0, D)
    Gamma_z = np.random.normal(0, D)

    # Componentes de la velocidad actual
    vx_i = vx[i]
    vy_i = vy[i]
    vz_i = vz[i]

    # Término de Lorentz: -q B × v
    F_Lorentz_x = q * B * vy_i
    F_Lorentz_y = -q * B * vx_i
    F_Lorentz_z = 0.0

    # Término de fricción
    F_fric_x = -beta * vx_i
    F_fric_y = -beta * vy_i
    F_fric_z = -beta * vz_i

    # Ecuación de Langevin
    dvx = (F_fric_x + F_Lorentz_x) * dt / m + Gamma_x
    dvy = (F_fric_y + F_Lorentz_y) * dt / m + Gamma_y
    dvz = (F_fric_z + F_Lorentz_z) * dt / m + Gamma_z

    # Actualizar velocidades
    vx[i+1] = vx_i + dvx
    vy[i+1] = vy_i + dvy
    vz[i+1] = vz_i + dvz
```

```python
plt.figure(figsize=(10, 5))
plt.plot(t, vx, label='$v_x$', color='red', linewidth=1, alpha=0.8)
plt.plot(t, vy, label='$v_y$', color='blue', linewidth=1, alpha=0.8)
plt.plot(t, vz, label='$v_z$', color='green', linewidth=1, alpha=0.8)
plt.xlabel('Tiempo')
plt.ylabel('Velocidad')
plt.title('Componentes de velocidad vs tiempo')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
